## Machine Learning - Predicting Survivors of the Titanic

```python
# Import the necessary packages used
import numpy as np
import pandas as pd
import os
```

#### Exploratory Data Analysis
Import the training & testing data and perform exploratory analysis on the data

```python
# Read in train data
train_data = pd.read_csv("./Raw_Data/train.csv")
train_data.head()
```

```python
# Analyze the train data
train_data.describe()
```

```python
# Read in the test data
test_data = pd.read_csv("./Raw_Data/test.csv")
test_data.head()
```

### Pattern Exploration
To begin, let's test a very basic pattern and hypothesize that all female passengers survived while all male passengers died. While this is a fairly aggressive hypothesis, the notion behind this is that the male spouse would let his female counterpart board the evacuation boats before he would. Before coding your model, it is often a good idea to sit down and think through the factors, or variables, that may have an impact on determining the outcome.

```python
# Determine % of women who survived
women = train_data.loc[train_data.Sex == 'female']['Survived']
rate_women = sum(women)/len(women)
print("Percentage of women who survived:", rate_women)
```

```python
# Determine % of men who survived
men = train_data.loc[train_data.Sex == 'male']['Survived']
rate_men = sum(men)/len(men)
print("Percentage of men who survived:", rate_men)
```

As you can see, it appears that roughly 75% of women survived whereas only 19% of men survived. A decent first try, but let's analyze the other variables in the dataset and determine the impact they have on predicting who survived.

### Random Forest Model
We will build what is known as a random forest model, constructing several "trees" that will determine whether the individual in the test data survived. 

```python
# Import the RandomForestClassifier module
from sklearn.ensemble import RandomForestClassifier
```

```python
# Instantiate y variable - what we are trying to predict
y = train_data["Survived"]

# Instantiate X variables for both train & test data - the factors we are using to predict y
# We will start with Pclass, Sex, SibSp, Parch
X = pd.get_dummies(train_data[['Pclass', 'Sex', 'SibSp', 'Parch']])
X_test = pd.get_dummies(test_data[['Pclass', 'Sex', 'SibSp', 'Parch']])
```

```python
# Create the RandomForestClassifier model and predict survival 
model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=1)
model.fit(X, y)
y_test = model.predict(X_test)
print(y_test)
print(sum(y_test))
```

```python
# Convert output to csv
output = pd.DataFrame({'PassengerId': test_data.PassengerId, 'Survived': y_test})
output.to_csv('submission.csv', index=False)
```
